```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = np.random.randn(20, 1)        # 20 data points, 1 feature
y = 3 * X.squeeze() + 2 + np.random.randn(20) * 0.5  
# true relationship: y = 3x + 2 (plus some noise)
```

```python
# Initialize weights randomly
weights = np.random.randn(1)
bias = np.random.randn(1)
learning_rate = 0.01
n = len(y)
losses = []
```

```python
# Training loop
for i in range(100):
    # Forward pass
    predictions = X @ weights + bias
    
    # Calculate MSE loss
    loss = np.mean((predictions - y) ** 2)
    losses.append(loss)
    
    # Compute gradients
    d_loss_weights = (2/n) * X.T @ (predictions - y)
    d_loss_bias = (2/n) * np.sum(predictions - y)
    
    # Update weights
    weights -= learning_rate * d_loss_weights
    bias -= learning_rate * d_loss_bias
    
    if i % 10 == 0:
        print(f"Iteration {i}: Loss = {loss:.4f}")
```

```python

# Plot loss curve
plt.plot(losses)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Loss vs Iterations')
plt.show()

print(f"Learned weights: {weights[0]:.2f} (true: 3.0)")
print(f"Learned bias: {bias[0]:.2f} (true: 2.0)")
```
